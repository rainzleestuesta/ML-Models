**Normalize the [Age] Column of pima_data since it does not match to the diabetes_data**

```python
pima_data.rename(columns={'Age': 'AGE'}, inplace=True)
```

**Check for null values in the columns [AGE] and [BMI] since it will be the keys to merge**

```python
diabetes_data[['AGE', 'BMI']].isnull().sum()
```

```python
diabetes_data.dropna(subset=['AGE', 'BMI'], inplace=True)
diabetes_data[['AGE', 'BMI']].isnull().sum()
```

**Print Datatypes**

```python
print(diabetes_data.dtypes)
print(pima_data.dtypes)
```

**As you can see, the [AGE] Column for the diabetes_data is in float thus it will cause error in merging. Thus, it should be converted to int**

```python
diabetes_data['AGE'] = diabetes_data['AGE'].astype(int)
```

**Drop rows with NaN in BMI or AGE before merging**

```python
diabetes_data.dropna(subset=['BMI', 'AGE'], inplace=True)
pima_data.dropna(subset=['BMI', 'AGE'], inplace=True)
```

**Merge the datasets**

```python
merged_data = pd.merge(diabetes_data, pima_data, on=['BMI', 'AGE'], how='outer')
merged_data
```

**Fill the null values in the datasets using KNN Imputer and their mean depending on the type of data**

Missing values before imputation:

```python
merged_data.isnull().sum()
```

```python
from sklearn.impute import KNNImputer

# Select all relevant numeric columns
features = merged_data.select_dtypes(include=['float64', 'int64'])

imputer = KNNImputer(n_neighbors=1)  # You can adjust n_neighbors as needed

# Apply KNN imputation
imputed_values = imputer.fit_transform(features)

# Create a new DataFrame with imputed values
imputed_df = pd.DataFrame(imputed_values, columns=features.columns)

# Replace original numeric columns with imputed ones
merged_data[features.columns] = imputed_df

merged_data.describe()
```

Missing values after imputation:

```python
merged_data.isnull().sum()
```

Drop Gender and Class since there is no way to fill in those type of data

```python
merged_data.drop(columns=['Gender', 'CLASS'], inplace=True)
```

**There are problem within the data, as you can see there are zero values within the columnns ['SkinThickness', 'Glucose', 'BloodPressure', 'Insulin', 'Outcome']. Which is considered invalid. Thus, filling them with the mean values would be the solution**

```python
columns_to_impute = ['SkinThickness', 'Glucose', 'BloodPressure', 'Insulin', 'Outcome']

# Iterate through each column and replace zeros with mean
for column in columns_to_impute:
    # Check for zero values
    zero_mask = merged_data[column] == 0
    print(f"Count of Zero Values in {column} before imputation:", zero_mask.sum())
    
    # Calculate mean or median of non-zero values
    non_zero_values = merged_data.loc[merged_data[column] != 0, column]
    
    # Check if there are any non-zero values before calculating mean/median
    if not non_zero_values.empty:
        mean_value = non_zero_values.mean()
        median_value = non_zero_values.median()
        merged_data.loc[zero_mask, column] = mean_value  # or use median_value
print()        
for column in columns_to_impute:
    print(f"Updated Count of Zero Values in {column} after imputation:", (merged_data[column] == 0).sum())

merged_data.describe()
```

**Remove outliers**

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
for col in columns_to_check:
    merged_data = remove_outliers(merged_data, col)
```

```python
merged_data.boxplot(column=columns_to_check)
plt.title('Boxplot After Outlier Removal')
plt.show()
```
